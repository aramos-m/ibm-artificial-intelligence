# Histogramas

```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import os

peliculas = pd.read_csv('RatingsPeliculas.csv')
peliculas.columns = ['Pelicula', 'Genero', 'RatingsCriticos', 'RatingsAudiencia', 'PresupuestoMillones', 'A単o']

peliculas.Pelicula = peliculas.Pelicula.astype('category')
peliculas.Genero = peliculas.Genero.astype('category')
peliculas.A単o = peliculas.A単o.astype('category')

peliculas.info()
```

---

```python
# M辿todo 1

# Diagrama 1
m1 = sns.distplot(peliculas.RatingsAudiencia, bins=15)
plt.show()
```

```python
# Diagrama 2
m2 = sns.distplot(peliculas.RatingsCriticos, bins=15)
plt.show()
```

```python
# M辿todo 2

# Diagrama 1
n1 = plt.hist(peliculas.RatingsAudiencia, bins=15)
plt.show()
```

```python
# Diagrama 2
n2 = plt.hist(peliculas.RatingsCriticos, bins=15)
plt.show()
```

# Histogramas Apilados

```python
#peliculas[peliculas.Genero == "Drama"].PresupuestoMillones

plt.hist(peliculas[peliculas.Genero =="Accion"].PresupuestoMillones, bins=15)
plt.hist(peliculas[peliculas.Genero =="Drama"].PresupuestoMillones, bins=15)
plt.hist(peliculas[peliculas.Genero =="Suspenso"].PresupuestoMillones, bins=15)
plt.show()
```

```python
plt.hist([peliculas[peliculas.Genero =="Accion"].PresupuestoMillones, \
          peliculas[peliculas.Genero =="Drama"].PresupuestoMillones, \
          peliculas[peliculas.Genero =="Suspenso"].PresupuestoMillones], \
          bins=15, stacked=True) # stacked: apila las barras

plt.show()
```

```python
for gen in peliculas.Genero.cat.categories:
    print(gen)
```

```python
lista1 = list()
mis_etiquetas = list()

for gen in peliculas.Genero.cat.categories:
    lista1.append(peliculas[peliculas.Genero==gen].PresupuestoMillones)
    mis_etiquetas.append(gen)
```

```python
h = plt.hist(lista1, bins=30, stacked=True, rwidth=1, label=mis_etiquetas) # rwidth ajusta el ancho relativo de las barras.
plt.legend()
plt.show()
```
