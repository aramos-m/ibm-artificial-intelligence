# Funciones para el análisis de datos


```python
def mi_grafico(lista_jugadores):
    for nombre in lista_jugadores:
        plt.plot(juegos[dic_jug[nombre]], c='Black', ls='--', marker='o', markersize=5, label=nombre)

    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.xticks(list(range(0,10)), temporadas, rotation='vertical')
    plt.show()
```

```python
mi_grafico(['KobeBryant', 'LeBronJames', 'DerrickRose'])
```

```python
def mi_grafico(lista_jugadores):
    col = {
        "KobeBryant":'Black', "JoeJohnson":'Red',
        "LeBronJames":'Green', "CarmeloAnthony":'Blue',
        "DwightHoward":'Purple',"ChrisBosh":'Orange',
        "ChrisPaul":'Brown',"KevinDurant":'Gray',
        "DerrickRose":'Pink',"DwayneWade":'Gold'
        }

    mar = {
        "KobeBryant":'o', "JoeJohnson":'s',
        "LeBronJames":'^', "CarmeloAnthony":'v',
        "DwightHoward":'D',"ChrisBosh":'>',
        "ChrisPaul":'<',"KevinDurant":'*',
        "DerrickRose":'h',"DwayneWade":'p'
        }
    
    for nombre in lista_jugadores:
        plt.plot(juegos[dic_jug[nombre]], c=col[nombre], ls='--', marker=mar[nombre], markersize=7, label=nombre)

    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.xticks(list(range(0,10)), temporadas, rotation='vertical')
    plt.show()
```

```python
def mi_grafico(matriz, lista_jugadores=jugadores):
    col = {
        "KobeBryant":'Black', "JoeJohnson":'Red',
        "LeBronJames":'Green', "CarmeloAnthony":'Blue',
        "DwightHoward":'Purple',"ChrisBosh":'Orange',
        "ChrisPaul":'Brown',"KevinDurant":'Gray',
        "DerrickRose":'Pink',"DwayneWade":'Gold'
        }

    mar = {
        "KobeBryant":'o', "JoeJohnson":'s',
        "LeBronJames":'^', "CarmeloAnthony":'v',
        "DwightHoward":'D',"ChrisBosh":'>',
        "ChrisPaul":'<',"KevinDurant":'*',
        "DerrickRose":'h',"DwayneWade":'p'
        }
    
    for nombre in lista_jugadores:
        plt.plot(matriz[dic_jug[nombre]], c=col[nombre], ls='--', marker=mar[nombre], markersize=7, label=nombre)

    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.xticks(list(range(0,10)), temporadas, rotation='vertical')
    plt.show()
```
