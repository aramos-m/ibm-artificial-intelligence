# Violinplot vs Boxplot

```python
import pandas as pd
import os

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

peliculas = pd.read_csv('RatingsPeliculas.csv')
peliculas.columns = ['Pelicula', 'Genero', 'RatingsCriticos', 'RatingsAudiencia', 'PresupuestoMillones', 'A単o']

peliculas.Pelicula = peliculas.Pelicula.astype('category')
peliculas.Genero = peliculas.Genero.astype('category')
peliculas.A単o = peliculas.A単o.astype('category')
peliculas.info()
```

---

```python
# Gr叩fico 1: Diagrama de viol鱈n.
# - Caja central: Muestra el 50% central de los datos (rango intercuart鱈lico), donde se encuentran la mayor鱈a de los valores.
# - Mediana: La l鱈nea en el centro de la caja, que indica el valor medio de los datos.
# - Bigotes: L鱈neas que indican los valores m叩s bajos y m叩s altos que no son at鱈picos.
# - Curvas: Muestran la densidad de los datos, es decir, cu叩ntos datos se agrupan en diferentes 叩reas. Las zonas m叩s anchas indican mayor concentraci坦n de datos.

z = sns.violinplot(x="Genero", y="RatingsCriticos", data=peliculas, palette="Set2")

plt.show()
```

```python
# Gr叩fica 2: Diagrama de caja.
# - Caja: Muestra el rango intercuart鱈lico (50% de los datos), desde el cuartil inferior (Q1) hasta el cuartil superior (Q3).
# - Mediana: La l鱈nea en el centro de la caja, que divide los datos en dos partes iguales.
# - Bigotes: L鱈neas que indican los valores m叩s bajos y m叩s altos dentro del rango de datos no at鱈picos.
# - Valores at鱈picos: Puntos fuera de los bigotes que representan datos que se alejan mucho del resto.

z = sns.boxplot(x="Genero", y="RatingsCriticos", data=peliculas, palette="Set2")

plt.show()
```

```python
# Gr叩fico 3: Diagrama de caja.

w = sns.boxplot(x="A単o", y="RatingsCriticos", data=peliculas[peliculas.Genero == 'Drama'], palette="Set2")

plt.show()
```

```python
# Gr叩fico 4: Diagrama de viol鱈n.

w = sns.violinplot(x="A単o", y="RatingsCriticos", data=peliculas[peliculas.Genero == 'Drama'], palette="Set2")

plt.show()
```
