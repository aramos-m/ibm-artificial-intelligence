# Seaborn

```python
# Importamos el m√≥dulo pandas
import pandas as pd

datos = pd.read_csv('DatosDemograficos.csv')

# Cambio de nombre a las columnas
datos.columns = ['NombrePais', 'CodigoPais', 'TasaNatalidad', 'UsuariosInternet', 'GrupoIngresos']
datos.head()

# Filas superiores
datos.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
datos.head()
```

```python
# Distribuci√≥n de los datos
vis1 = sns.distplot(datos['UsuariosInternet'])
```

```python
# Distribuci√≥n de los datos
# Visualizaci√≥n 1: Gr√°fico de densidad kernel (KDE) con relleno.
# Esta gr√°fica muestra una estimaci√≥n de la distribuci√≥n de probabilidad de los datos, usando la t√©cnica de estimaci√≥n de densidad kernel (KDE).
# Par√°metros:
# - 'kind'='kde': Indica que queremos un gr√°fico de densidad en lugar de un histograma.
# - 'color'='g': El color de la curva es verde ('g' es abreviatura de green).
# - 'fill'=True: La curva de densidad est√° rellena bajo la curva.
vis1 = sns.displot(datos['UsuariosInternet'], kind='kde', color='g', fill=True)
plt.show()
```

```python
# Distribuci√≥n de los datos
# Visualizaci√≥n 2: Histograma con curva KDE superpuesta.
# Esta gr√°fica muestra un histograma con 20 divisiones (bins), lo cual es √∫til para ver la distribuci√≥n de frecuencia de los valores.
# Adem√°s, se a√±ade una curva KDE (Estimaci√≥n de Densidad Kernel) superpuesta para mostrar la distribuci√≥n suave de los datos.
# Par√°metros:
# - 'kde'=True: A√±ade la curva KDE al histograma.
# - 'bins'=20: El n√∫mero de divisiones (bins) en el histograma es 20.
vis1 = sns.histplot(datos['UsuariosInternet'], kde=True, bins=20)
plt.show()
```

```python
vis2 = sns.boxplot(data=datos, x='GrupoIngresos', y='TasaNatalidad')
plt.show()
```
