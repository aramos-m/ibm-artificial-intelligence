# Seaborn

```python
# Importamos el módulo pandas
import pandas as pd

datos = pd.read_csv('DatosDemograficos.csv')

# Cambio de nombre a las columnas
datos.columns = ['NombrePais', 'CodigoPais', 'TasaNatalidad', 'UsuariosInternet', 'GrupoIngresos']
datos.head()

# Filas superiores
datos.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
datos.head()
```

```python
# Distribución de los datos
vis1 = sns.displot(datos['UsuariosInternet'])
plt.show()
```

```python
# Distribución de los datos
# Visualización 1: Gráfico de densidad kernel (KDE) con relleno.
# Esta gráfica muestra una estimación de la distribución de probabilidad de los datos, usando la técnica de estimación de densidad kernel (KDE).
# Parámetros:
# - 'kind'='kde': Indica que queremos un gráfico de densidad en lugar de un histograma.
# - 'color'='g': El color de la curva es verde ('g' es abreviatura de green).
# - 'fill'=True: La curva de densidad está rellena bajo la curva.
vis1 = sns.displot(datos['UsuariosInternet'], kind='kde', color='g', fill=True)
plt.show()
```

```python
# Distribución de los datos
# Visualización 2: Histograma con curva KDE superpuesta.
# Esta gráfica muestra un histograma con 20 divisiones (bins), lo cual es útil para ver la distribución de frecuencia de los valores.
# Además, se añade una curva KDE (Estimación de Densidad Kernel) superpuesta para mostrar la distribución suave de los datos.
# Parámetros:
# - 'kde'=True: Añade la curva KDE al histograma.
# - 'bins'=20: El número de divisiones (bins) en el histograma es 20.
vis1 = sns.histplot(datos['UsuariosInternet'], kde=True, bins=20)
plt.show()
```

```python
vis2 = sns.boxplot(data=datos, x='GrupoIngresos', y='TasaNatalidad')
plt.show()
```

---


```python
# Más visualizaciones con seaborn

vis3 = sns.lmplot(data=datos, x='UsuariosInternet', y='TasaNatalidad', \
                  fit_reg=False, hue='GrupoIngresos')
plt.show()
```

```python
vis3 = sns.lmplot(data=datos, x='UsuariosInternet', y='TasaNatalidad', \
                  fit_reg=False, hue='GrupoIngresos', height=5, \
                  scatter_kws={'s': 100}) # Tamaño de los puntos
plt.show()
```
