# Operaciones Básicas con DataFrames

```python
# Importamos el módulo pandas
import pandas as pd

datos = pd.read_csv('DatosDemograficos.csv')

# Filas superiores
datos.head(10)
```

```python
# Cambio de nombre a las columnas
datos.columns = ['NombrePais', 'CodigoPais', 'TasaNatalidad', 'UsuariosInternet', 'GrupoIngresos']
datos.head()
```

```python
# Operaciones matemáticas

# No tiene sentido, pero es una prueba
resultado = datos.TasaNatalidad * datos.UsuariosInternet
resultado.head()
```

```python
# Agregar una columna
datos['MiCalculo'] = datos.TasaNatalidad * datos.UsuariosInternet
datos.head()
```

```python
# Borrar una columna
datos.head()
```

```python
datos = datos.drop('MiCalculo', axis=1) # Necesitamos sobreescribir el dataframe
                                        # axis=0 para filas, axis=1 para columnas
```

```python
datos.head(10)
```

# Filtrado de DataFrame

```python
# El filtrado se aplica a las filas
datos.head(10)
```

```python
datos.UsuariosInternet < 2
```

```python
datos[filtro] # Hacemos el slicing con el filtro
```

```python
filtro2 = datos.TasaNatalidad > 40
datos[filtro2]
```

```python
# Podemos combinar filtros.

# datos[datos.TasaNatalidad > 40 and datos.Usuarios < 2] # Error 
datos[filtro & filtro2] # & aplica los filtros elemento a elemento
```

# Acceso a elementos individuales de un DataFrame

```python
# Usando .at() y .iat()

# datos[2,2] #Error. No se puede acceder a un elemento de un DataFrame con un índice
datos.iat[3,4] # para índices, ubicación con números enteros 
```

```python
datos.at[2, 'TasaNatalidad'] # para etiquetas
```

```python
sub10 = datos [::10]
sub10
```

```python
sub10.iat[10,0] # Cuenta el 10º elemento del nuevo subconjunto
```

```python
sub10.at[10, 'NombrePais'] # Busca la etiqueta, por eso tenemos resultado diferentes.
```
