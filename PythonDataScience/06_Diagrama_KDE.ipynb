# Diagrama KDE

```python
import pandas as pd
import os

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

peliculas = pd.read_csv('RatingsPeliculas.csv')
peliculas.columns = ['Pelicula', 'Genero', 'RatingsCriticos', 'RatingsAudiencia', 'PresupuestoMillones', 'A単o']

peliculas.Pelicula = peliculas.Pelicula.astype('category')
peliculas.Genero = peliculas.Genero.astype('category')
peliculas.A単o = peliculas.A単o.astype('category')
peliculas.info()
```

---

```python
visualizacion1 = sns.lmplot(x="RatingsCriticos", y="RatingsAudiencia", 
                            data=peliculas, fit_reg=False, hue="Genero", 
                            scatter_kws={"s": 50}, height=5, aspect=1)
plt.show()
```

```python
# Diagrama de Kernel Density
# Definici坦n de Kernel Density: https://en.wikipedia.org/wiki/Kernel_density_estimation

sns.set_style("darkgrid") # Tema oscuro
k1 = sns.kdeplot(x=peliculas.RatingsCriticos, y=peliculas.RatingsAudiencia, \
                 shade=True, shade_lowest=False, cmap="Purples")

# Tip para que se vea mejor los bordes.
k2 = sns.kdeplot(x=peliculas.RatingsCriticos, y=peliculas.RatingsAudiencia, \
                 cmap="BuPu_r")
plt.show()
```

```python
# KDE con mapa de calor en tonos rojos. "Reds" es un colormap en tonos c叩lidos.
k3 = sns.kdeplot(x=peliculas.RatingsCriticos, y=peliculas.RatingsAudiencia, 
                 shade=True, shade_lowest=False, cmap="Reds", 
                 alpha=0.7)  # alpha ajusta la transparencia (0.0 a 1.0)
plt.show()
```
