```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importamos funciones de plotly para crear gr谩ficos interactivos
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

import warnings
warnings.filterwarnings("ignore")
```

```python
%matplotlib inline 
# Mostrar gr谩ficos en la consola
```

```python
dataframe = pd.DataFrame(np.random.randn(100,4), columns=['a','b','c','d'])
dataframe.head()
```

```python
dataframe.plot()
plt.show()
```

```python
# Genera un gr谩fico interactivo de l铆neas
dataframe.iplot()
```

```python
# Crea un gr谩fico de dispersi贸n con marcadores
dataframe.iplot(kind='scatter', x='a', y='b', mode='markers')
```

```python
# Suma las columnas y crea un gr谩fico de barras con el resultado
dataframe.sum().iplot(kind='bar')
```

```python
# Crea un gr谩fico de caja (box plot)
dataframe.iplot(kind='box')
```

```python
# Crea un histograma de la columna 'a' con 30 bins
dataframe['a'].iplot(kind='hist', bins=30)
```

```python
# Crea un gr谩fico de dispersi贸n entre las columnas 'a' y 'b'
dataframe[['a','b']].iplot(kind='spread')
```

```python
# Crea un gr谩fico de burbujas con las columnas 'a' y 'b' como ejes, 'c' para el tama帽o de las burbujas
dataframe.iplot(kind='bubble', x='a', y='b', size='c', color='mediumorchid')
```
