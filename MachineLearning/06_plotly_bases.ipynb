## Sección 8: Módulos Plotly y Cufflinks para análisis de datos

<details>
<summary><b>📊 PLOTLY & CUFFLINKS</b></summary>

- **Plotly**: Una librería de Python para crear gráficos interactivos y visualizarlos en línea o localmente.  
  - Instalación: `pip install plotly`
<br></br>
- **Cufflinks**: Una extensión que conecta **Pandas** con **Plotly**, facilitando la creación de gráficos directamente desde DataFrames.  
  - Instalación: `pip install cufflinks`

</details>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importamos funciones de plotly para crear gráficos interactivos
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore")
```

```python
# Inicializa el modo interactivo de Plotly en Jupyter
init_notebook_mode(connected=True)
```

```python
cf.go_offline() # Asegurar visualización offline
```

```python
%matplotlib inline 
# Mostrar gráficos en la consola
```

```python
dataframe = pd.DataFrame(np.random.randn(100,4), columns=['a','b','c','d'])
dataframe.head()
```

```python
dataframe.plot()
plt.show()
```

```python
# Genera un gráfico interactivo de líneas
figura = dataframe.iplot(asFigure=True)
figura.write_html("line-iplot.html")
```

<a href="iplot-graphs/line-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
# Crea un gráfico de dispersión con marcadores
figura = dataframe.iplot(kind='scatter', x='a', y='b', mode='markers', asFigure=True)
figura.write_html("scatter-iplot.html")
```

<a href="iplot-graphs/scatter-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
# Suma las columnas y crea un gráfico de barras con el resultado
figura = dataframe.sum().iplot(kind='bar', asFigure=True)
figura.write_html("bar-iplot.html")
```

<a href="iplot-graphs/bar-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
# Crea un gráfico de caja (box plot)
figura = dataframe.iplot(kind='box', asFigure=True)
figura.write_html("box-iplot.html")
```

<a href="iplot-graphs/box-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
# Crea un histograma de la columna 'a' con 30 bins
figura = dataframe['a'].iplot(kind='hist', bins=30, asFigure=True)
figura.write_html("hist-iplot.html")
```

<a href="iplot-graphs/hist-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
# Crea un gráfico de dispersión entre las columnas 'a' y 'b'
figura = dataframe[['a','b']].iplot(kind='spread', asFigure=True)
figura.write_html("spread-iplot.html")
```

<a href="iplot-graphs/spread-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
# Crea un gráfico de burbujas con las columnas 'a' y 'b' como ejes, 'c' para el tamaño de las burbujas
figura = dataframe.iplot(kind='bubble', x='a', y='b', size='c', color='mediumorchid', asFigure=True)
figura.write_html("bubble-iplot.html")
```

<a href="iplot-graphs/bubble-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
dataframe2 = pd.DataFrame({'a':[1,2,3,4], 'b':[30,40,20,10], 'c':[12,16,18,15]})
dataframe2
```

```python
# Crea un gráfico 3D de superficie
figura = dataframe2.iplot(kind='surface', asFigure=True)
figura.write_html("surface-iplot.html")
```

<a href="iplot-graphs/surface-iplot.html" target="_blank" style="display:inline-block; background-color:#4CAF50; color:white; padding:10px 20px; text-align:center; text-decoration:none; border-radius:5px;">Ver gráfico interactivo</a>

```python
# Crea un gráfico 3D de dispersión
figura = dataframe2.iplot(kind='scatter3d', x='a', y='b', z='c', asFigure=True)
figura.write_html("scatter3d-iplot.html")
```
