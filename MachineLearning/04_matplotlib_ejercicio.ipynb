- Crear una variable "x" que tenga 10 valores equidistantes entre los valores 0 y 20
- Crear una variable "y" que tenga los valores de la funcion "x al cuadrado + 5"

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,20,10)
x
```

```python
y = (x**2) + 5
y
```

```python
plt.plot(x,y,'r',linewidth=5,linestyle=':')
```
