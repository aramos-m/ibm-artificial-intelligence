### Ejercicio algoritmo de las k-medias

````python
import pandas as pd
import numpy as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Importar funci√≥n para crear datos en grupos o clusters
from sklearn.datasets import make_blobs
````

````python
# Crear 200 puntos con 2 caracter√≠sticas (axis x y axis y) distribuidos en 4 grupos
datos = make_blobs(n_samples=200, n_features=2, centers=4)
````

````python
datos[0] # Mostrar las coordenadas de los puntos generados
````

````python
datos[1] # Mostrar las etiquetas reales que identifican a qu√© grupo pertenece cada punto
````

````python
plt.scatter(datos[0][:,0], datos[0][:,1])
````

````python
from sklearn.cluster import KMeans
````

````python
modelo = KMeans(n_clusters=4)
````

````python
modelo.fit(datos[0])
````

````python
modelo.cluster_centers_ # Obtener las coordenadas de los centroides calculados por el modelo
````

````python
modelo.labels_ # Mostrar las etiquetas asignadas por el modelo a cada punto
````

````python
datos[1] # Mostrar las etiquetas reales de los puntos generados (grupos originales)
````

````python
# Crear dos gr√°ficos para comparar los grupos asignados por el modelo con los originales
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(12,4))

ax1.scatter(datos[0][:,0], datos[0][:,1], c=modelo.labels_)
ax1.set_title("Algoritmo de k-medias")

ax2.scatter(datos[0][:,0], datos[0][:,1], c=datos[1])
ax2.set_title("Datos originales")
````
