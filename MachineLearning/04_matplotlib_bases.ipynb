## Secci칩n 6: M칩dulo Matplotlib para an치lisis de datos

```python
import matplotlib.pyplot as plt
import numpy as np
```

---
### Aplicar POO a la visualizaci칩n de datos

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y,'lightgreen')
plt.title("Este es mi gr치fico")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
# plt.show() En Jupyter Notebook no es necesario
```

```python
plt.subplot(1,2,1) # 1 columna, 2 filas
plt.plot(x,y,'g')
plt.subplot(1,2,2) # 1 columna, 2 filas
plt.plot(y,x,'blue')
```

```python
plt.subplot(2,1,1) # 2 columnas, 1 fila
plt.plot(x,y,'green')
plt.subplot(2,1,2) # 2 columnas, 1 fila
plt.plot(y,x,'blue')
```

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'green')
```

```python
figura = plt.figure() # Crear el "lienzo" donde dibujar el gr치fico
grafico1 = figura.add_axes([0.1,0.1,0.8,0.8]) # define la posici칩n y el tama침o respecto al lienzo [izquierda, abajo, ancho, alto]
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.2,0.5,0.3,0.3])
grafico2.plot(y,x,'blue')
```

---
### Multigr치ficos (filas, columnas)

```python
figura, graficos = plt.subplots(nrows=2,ncols=2) # Crear figura y una cuadr칤cula de 2x2 gr치ficos
```

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)

graficos[0][0].plot(x,y,'blue')
graficos[0][0].set_title("Azul")

graficos[0][1].plot(x,y,'red')
graficos[0][1].set_title("Rojo")

graficos[1][0].plot(x,y,'y')
graficos[1][0].set_title("Amarillo")

graficos[1][1].plot(x,y,'g')
graficos[1][1].set_title("Verde")
```

---
### Tama침o del gr치fico

```python
figura = plt.figure(figsize=(10,2)) # Establece el tama침o de la figura (ancho, alto)
grafico = figura.add_axes([0.1, 0.1, 0.9, 0.9])
grafico.plot(x,y,'r')
```

---
### Dos gr치ficos en la misma figura

```python
figura = plt.figure()

graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, label='Al cuadrado')
graficos.plot(x, x**3, label='Al cubo') # Ambos gr치ficos comparten el mismo espacio de coordenadas (ejes)
graficos.legend(loc=0) # loc=0 para ubicar la leyenda en la posici칩n 칩ptima
```

---
### Color del gr치fico

```python
figura = plt.figure()

grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,x**2,'purple')
grafico.plot(x,x**3,'#a4ccee') # Color hexadecimal personalizado
```

---
### Linea del gr치fico (grosor, transparencia y tipo)

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'r',linewidth=4, alpha=0.4, linestyle='-.') # Graficar con grosor 5, transparencia 0.4 y estilo de l칤nea discontinua

# M치s tipos de l칤nea: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py
```
