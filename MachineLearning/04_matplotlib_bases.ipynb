## Sección 6: Módulo Matplotlib para análisis de datos

```python
import matplotlib.pyplot as plt
import numpy as np
```

---
### Aplicar POO a la visualización de datos

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y,'lightgreen')
plt.title("Este es mi gráfico")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
# plt.show() En Jupyter Notebook no es necesario
```

```python
plt.subplot(1,2,1) # 1 columna, 2 filas
plt.plot(x,y,'g')
plt.subplot(1,2,2) # 1 columna, 2 filas
plt.plot(y,x,'blue')
```

```python
plt.subplot(2,1,1) # 2 columnas, 1 fila
plt.plot(x,y,'green')
plt.subplot(2,1,2) # 2 columnas, 1 fila
plt.plot(y,x,'blue')
```

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'green')
```

```python
figura = plt.figure() # Crear el "lienzo" donde dibujar el gráfico
grafico1 = figura.add_axes([0.1,0.1,0.8,0.8]) # define la posición y el tamaño respecto al lienzo [izquierda, abajo, ancho, alto]
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.2,0.5,0.3,0.3])
grafico2.plot(y,x,'blue')
```

---
### Multigráficos (filas, columnas)

```python
figura, graficos = plt.subplots(nrows=2,ncols=2) # Crear figura y una cuadrícula de 2x2 gráficos
```

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)

graficos[0][0].plot(x,y,'blue')
graficos[0][0].set_title("Azul")

graficos[0][1].plot(x,y,'red')
graficos[0][1].set_title("Rojo")

graficos[1][0].plot(x,y,'y')
graficos[1][0].set_title("Amarillo")

graficos[1][1].plot(x,y,'g')
graficos[1][1].set_title("Verde")
```

---
### Tamaño del gráfico

```python
figura = plt.figure(figsize=(10,2)) # Establece el tamaño de la figura (ancho, alto)
grafico = figura.add_axes([0.1, 0.1, 0.9, 0.9])
grafico.plot(x,y,'r')
```

---
### Dos gráficos en la misma figura

```python
figura = plt.figure()

graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, label='Al cuadrado')
graficos.plot(x, x**3, label='Al cubo') # Ambos gráficos comparten el mismo espacio de coordenadas (ejes)
graficos.legend(loc=0) # loc=0 para ubicar la leyenda en la posición óptima
```

---
### Color del gráfico

```python
figura = plt.figure()

grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,x**2,'purple')
grafico.plot(x,x**3,'#a4ccee') # Color hexadecimal personalizado
```

---
### Linea del gráfico (grosor, transparencia y tipo)

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'r',linewidth=4, alpha=0.4, linestyle='-.') # Graficar con grosor 5, transparencia 0.4 y estilo de línea discontinua

# Más tipos de línea: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py
```
